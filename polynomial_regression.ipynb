```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/chentingkao/PracticeData/Position_Salaries.csv')
df.info()
```

```python
# 設定一個範圍 return DataFrame
feature_df = df.iloc[:, 1:2]

# 單選一個 return series
target = df.iloc[:, 2]

from sklearn.model_selection import train_test_split

# 不管是 DataFrame 或是 Series 都能 split
X_train, X_test, y_train, y_test = train_test_split(feature_df, target, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt

def draw_linear_regression(predictions):
  plt.scatter(feature_df, target, color = 'red')
  plt.plot(feature_df, predictions, color = 'blue')
  plt.title('Level vs Salary')
  plt.xlabel('Level')
  plt.ylabel('Salary')
  plt.show()
  return
```

```python
draw_linear_regression(linear_regression.predict(feature_df))
```

```python
# Polynomial Regression in sklearn is done through data preprocessing
from sklearn.preprocessing import PolynomialFeatures

# the key is to transform the features into poly features
poly_processing = PolynomialFeatures(degree = 4)
poly_feature_df = poly_processing.fit_transform(feature_df)

# it is still using LinearRegression
poly_regression = LinearRegression()
poly_regression.fit(poly_feature_df, target)
```

```python
draw_linear_regression(poly_regression.predict(poly_feature_df))
```
