```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/chentingkao/PracticeData/titanic.csv')
df.head()
```

```python
# know your data
pd.info()
```

```python
df.describe()
```

```python
# Create another feature to tell if the person travels alone.
# the return type np.where is numpy.ndarray
# that also means Pandas DataFrame can turn a 2-d ndarray into a column
df['Travelalone'] = np.where((df['SibSp'] + df['Parch']) == 0, 1, 0)
```

```python
# drop unnecessary columns
df_after_drop = df.drop(['PassengerId', 'Ticket', 'Name', 'Cabin', 'Parch', 'SibSp'], axis = 1)
```

```python
# find invalid value
df_after_drop.isna().sum()
```

```python
# fill Na value with median of the column
df_after_drop['Age'].fillna(df_after_drop['Age'].median(skipna = True), inplace = True)
```

```python
# Turn categorical values into one-hot encoding
# drop_first=True:
# This argument indicates that the first category in each categorical variable
# should be dropped. This is often done to avoid multicollinearity in regression
# models, as one category can be inferred from the others.

df_one_hot = pd.get_dummies(df_after_drop, columns = ['Pclass', 'Sex', 'Embarked'], drop_first = True)
df_one_hot.head()
```

```python
# Prepare feature (x) and target (y)
feature_df = df_one_hot.drop(['Survived'], axis = 1)
target_df = df_one_hot['Survived']
```

```python
# initializer scalers to standardize data
from sklearn.preprocessing import MinMaxScaler, StandardScaler

mm_scaler = MinMaxScaler()
std_scaler = StandardScaler()
```

```python
# fit_transform returns "numpy ndarray"
feature_mm = mm_scaler.fit_transform(feature_df)
feature_df_mm = pd.DataFrame(feature_mm, columns = feature_df.columns)
feature_df_mm.head()
```

```python
feature_std = std_scaler.fit_transform(feature_df)
feature_df_std = pd.DataFrame(feature_std, columns = feature_df.columns)
feature_df_std.head()
```
